### Store the datasets in database tables

To analyze the data using SQL, it first needs to be loaded into SQLite DB.
We will create three tables in as under:

1.  **CENSUS_DATA**
2.  **CHICAGO_PUBLIC_SCHOOLS**
3.  **CHICAGO_CRIME_DATA**


Load the `pandas` and `sqlite3` libraries and establish a connection to `FinalDB.db`


```python
# -----------------------------------------------------------------
# Import the libraries we need
# -----------------------------------------------------------------
import pandas as pd                # for loading CSVs and writing to the database
import sqlite3                     # Python's built-in SQLite DB-API driver
from IPython.display import display  # for nicely displaying jupysql result tables

# -----------------------------------------------------------------
# Create the TRADITIONAL sqlite3 connection object
# -----------------------------------------------------------------
# This is the same "Connection Object" we covered earlier:
#   - it opens (or creates) the file FinalDB.db
#   - it will be used later for .cursor(), .execute(), .commit(), etc.
conn = sqlite3.connect('FinalDB.db')

print("Connected to FinalDB.db using sqlite3 ->", conn)
```

Load the SQL magic module


```python
# -----------------------------------------------------------------
# Load the ipython-sql / jupysql extension
# -----------------------------------------------------------------
# This gives the notebook access to the %sql (line) and %%sql (cell) magics.
# If you see "The sql extension is already loaded. To reload it, use:
# %reload_ext sql" -- that is NOT an error, it just means it was already
# loaded in a previous cell run.
%load_ext sql
```

Use `Pandas` to load the data available in the links above to dataframes. Use these dataframes to load data on to the database `FinalDB.db` as required tables.


```python
# -----------------------------------------------------------------
# Step 1: Load each CSV directly from its URL into a pandas DataFrame
# -----------------------------------------------------------------
# pandas can read a CSV straight from a web URL -- no need to download
# the file manually first.

census_data = pd.read_csv('../Data/ChicagoCensusData.csv')
schools_data = pd.read_csv('../Data/ChicagoPublicSchools.csv')
crime_data = pd.read_csv('../Data/ChicagoCrimeData.csv')

# -----------------------------------------------------------------
# Step 2: Write each DataFrame into FinalDB.db as a table
# -----------------------------------------------------------------
# .to_sql(table_name, connection, if_exists, index)

#   table_name  -> name of the table to create inside FinalDB.db
#   conn        -> the sqlite3 connection object created above
#   if_exists='replace' -> if the table already exists, drop & recreate it
#   index=False -> don't add pandas' row-number column (0,1,2,...) as an extra column in the table

n1 = census_data.to_sql("CENSUS_DATA", conn, if_exists='replace', index=False)
n2 = schools_data.to_sql("CHICAGO_PUBLIC_SCHOOLS", conn, if_exists='replace', index=False)
n3 = crime_data.to_sql("CHICAGO_CRIME_DATA", conn, if_exists='replace', index=False)

print("CENSUS_DATA rows written:", n1)
print("CHICAGO_PUBLIC_SCHOOLS rows written:", n2)
print("CHICAGO_CRIME_DATA rows written:", n3)

# Since pandas 1.4+, .to_sql() returns the number of rows it wrote to the table. So this tells you:
```

```python
# magic based equivalent to .to_sql()
%sql --persist census_data
%sql --persist schools_data
%sql --persist crime_data

# Your original code creates tables named CENSUS_DATA, CHICAGO_PUBLIC_SCHOOLS, and CHICAGO_CRIME_DATA 
# (uppercase, specific names) — but --persist always uses the DataFrame's variable name as the table name. 
# You can't rename it directly in the --persist call.

# trick
CENSUS_DATA = census_data
%sql --persist CENSUS_DATA

CHICAGO_PUBLIC_SCHOOLS = schools_data
%sql --persist CHICAGO_PUBLIC_SCHOOLS

CHICAGO_CRIME_DATA = crime_data
%sql --persist CHICAGO_CRIME_DATA

# Handling "table already exists" errors
%sql --persist --force CENSUS_DATA      # newer jupysql: overwrite if exists

# or manually drop existing table and then insert
%sql DROP TABLE IF EXISTS CENSUS_DATA
%sql --persist CENSUS_DATA
```

Establish a connection between SQL magic module and the database `FinalDB.db`


```python
# -----------------------------------------------------------------
# Connect the %sql / %%sql magic to the SAME database file
# -----------------------------------------------------------------
# Connection string format:  sqlite:///<filename>
#   sqlite://   -> protocol separator (always "://")
#   /FinalDB.db -> path to the database file (one extra "/")
#


# NOTE: this is a SEPARATE connection from "conn" above. Both can be
# open to the same FinalDB.db file at the same time -- "conn" is used
# by pandas/sqlite3 code, and this one is used by %sql / %%sql.
%sql sqlite:///FinalDB.db
```

### Problem 1

```python
# ===================================================================
# METHOD 1: jupysql (%sql magic)
# ===================================================================
# %sql sends the SQL statement to the database connected in Cell 10
# and returns a "ResultSet" object that we store in result_sql.
result_sql = %sql SELECT COUNT(*) AS Total_Crimes FROM CHICAGO_CRIME_DATA
display(result_sql)
```

### Problem 3

```python
# ===================================================================
# METHOD 1: jupysql (%sql magic)
# ===================================================================
# LIKE '%MINOR%' -> matches any DESCRIPTION that CONTAINS the word
# "MINOR" anywhere in the text. The % symbols are wildcards meaning
# "anything can come before/after this word".
result_sql = %sql SELECT CASE_NUMBER, DESCRIPTION FROM CHICAGO_CRIME_DATA WHERE DESCRIPTION LIKE '%MINOR%';
display(result_sql)

# ===================================================================
# METHOD 2: Traditional sqlite3 (connection + cursor)
# ===================================================================
cursor = conn.cursor()
cursor.execute("""
    SELECT CASE_NUMBER, DESCRIPTION
    FROM CHICAGO_CRIME_DATA
    WHERE DESCRIPTION LIKE '%MINOR%'
""")
rows = cursor.fetchall()
for row in rows:
    print(row)
```

### Problem 6

##### List the type of schools along with the average safety score for each type.


```python
%sql pragma table_info([CHICAGO_PUBLIC_SCHOOLS])
```

```python
# ===================================================================
# METHOD 1: jupysql (%sql magic)
# ===================================================================
# The column name "Elementary, Middle, or High School" contains
# SPACES and COMMAS, so it MUST be wrapped in double quotes "..."
# (covered earlier: "Querying column names with spaces").
#
# AVG(SAFETY_SCORE)        -> average safety score per group
# GROUP BY "Elementary..." -> one row PER school type (Elementary/Middle/High)
result_sql = %sql SELECT "Elementary, Middle, or High School" AS School_Type, AVG(SAFETY_SCORE) AS Average_Safety_Score FROM CHICAGO_PUBLIC_SCHOOLS GROUP BY School_Type
display(result_sql)

# ===================================================================
# METHOD 2: Traditional sqlite3 (connection + cursor)
# ===================================================================
cursor = conn.cursor()
cursor.execute("""
    SELECT "Elementary, Middle, or High School" AS School_Type,
           AVG(SAFETY_SCORE) AS Average_Safety_Score
    FROM CHICAGO_PUBLIC_SCHOOLS
    GROUP BY School_Type
""")
rows = cursor.fetchall()
for row in rows:
    print(row)
```

### Problem 8

##### Which community area is most crime prone? Display the coumminty area number only.


```python
# ===================================================================
# METHOD 1: jupysql (%sql magic)
# ===================================================================
# GROUP BY COMMUNITY_AREA_NUMBER -> one row per community area
# COUNT(*)                       -> how many crime records that area has
# ORDER BY ... DESC + LIMIT 1    -> keep only the area with the MOST crimes
result_sql = %sql SELECT COMMUNITY_AREA_NUMBER, COUNT(*) AS Number_of_Crimes FROM CHICAGO_CRIME_DATA GROUP BY COMMUNITY_AREA_NUMBER ORDER BY Number_of_Crimes DESC LIMIT 1;
display(result_sql)

# ===================================================================
# METHOD 2: Traditional sqlite3 (connection + cursor)
# ===================================================================
cursor = conn.cursor()
cursor.execute("""
    SELECT COMMUNITY_AREA_NUMBER, COUNT(*) AS Number_of_Crimes
    FROM CHICAGO_CRIME_DATA
    GROUP BY COMMUNITY_AREA_NUMBER
    ORDER BY Number_of_Crimes DESC
    LIMIT 1
""")
row = cursor.fetchone()
print("Most crime-prone Community Area Number:", row[0])
print("(Number of crimes recorded there:", row[1], ")")
```

### Problem 9

##### Use a sub-query to find the name of the community area with highest hardship index


```python
# ===================================================================
# METHOD 1: jupysql (%sql magic)
# ===================================================================
# The part inside (...) is the SUB-QUERY -- it runs FIRST and returns
# a single value: the highest HARDSHIP_INDEX in the whole table.
# The outer query then finds the row whose HARDSHIP_INDEX equals
# that value.
result_sql = %sql SELECT COMMUNITY_AREA_NAME, HARDSHIP_INDEX FROM CENSUS_DATA WHERE HARDSHIP_INDEX = (SELECT MAX(HARDSHIP_INDEX) FROM CENSUS_DATA);
display(result_sql)

# ===================================================================
# METHOD 2: Traditional sqlite3 (connection + cursor)
# ===================================================================
cursor = conn.cursor()
cursor.execute("""
    SELECT COMMUNITY_AREA_NAME, HARDSHIP_INDEX
    FROM CENSUS_DATA
    WHERE HARDSHIP_INDEX = (SELECT MAX(HARDSHIP_INDEX) FROM CENSUS_DATA)
""")
row = cursor.fetchone()
print("Community area with the highest hardship index:", row[0])
print("Hardship Index value:", row[1])
```

### Problem 10

##### Use a sub-query to determine the Community Area Name with most number of crimes?


```python
# ===================================================================
# METHOD 1: jupysql (%sql magic)
# ===================================================================
# Inner sub-query (runs first):
#   - groups CHICAGO_CRIME_DATA by COMMUNITY_AREA_NUMBER
#   - counts crimes per area, sorts descending, keeps the TOP 1
#   -> returns the COMMUNITY_AREA_NUMBER with the most crimes
#
# Outer query:
#   - looks up that COMMUNITY_AREA_NUMBER in CENSUS_DATA
#   - returns its COMMUNITY_AREA_NAME
result_sql = %sql SELECT COMMUNITY_AREA_NAME FROM CENSUS_DATA WHERE COMMUNITY_AREA_NUMBER = (SELECT COMMUNITY_AREA_NUMBER FROM CHICAGO_CRIME_DATA GROUP BY COMMUNITY_AREA_NUMBER ORDER BY COUNT(*) DESC LIMIT 1);
display(result_sql)

# ===================================================================
# METHOD 2: Traditional sqlite3 (connection + cursor)
# ===================================================================
cursor = conn.cursor()
cursor.execute("""
    SELECT COMMUNITY_AREA_NAME
    FROM CENSUS_DATA
    WHERE COMMUNITY_AREA_NUMBER = (
        SELECT COMMUNITY_AREA_NUMBER
        FROM CHICAGO_CRIME_DATA
        GROUP BY COMMUNITY_AREA_NUMBER
        ORDER BY COUNT(*) DESC
        LIMIT 1
    )
""")
row = cursor.fetchone()
print("Community area with the most crimes:", row[0])
```
